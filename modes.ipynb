```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 參數
L = 10
a = 3

# 使用 Geogebra找 y = (ℯ^(x)+ℯ^(-x))/2 + 1/cos(x) 的根
root = [1.87510, 
        4.69409,
        7.85476,
        10.99554]

# root = sqrt(omega)*L/a 求 eigen = sqrt(omega)/a 
eigen = []
omega = []
for r in root:
   eigen.append(r/L)
   omega.append((r*(a/L))**2)
```

```python
def mode(x, eigen, A=100, B=1):
    return A * np.sin(eigen * x) + B * np.cosh(eigen * x)    # particular F(x) 
```

```python
x_axis = np.linspace(0, L, 10000)  # x取值

# 繪製四個 modes
plt.figure(figsize=(12, 8))
for i in range(4):
    l = eigen[i]
    y_axis = mode(x_axis, l)
    plt.plot(x_axis, y_axis, label=f'mode {i + 1}, ω ≈ {omega[i]:.2f}')
    
plt.xlabel('x')
plt.ylabel('displacement')
plt.title('normal mode')
plt.legend()
plt.grid(True)
plt.show()
```
